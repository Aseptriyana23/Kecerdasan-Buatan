```python
import matplotlib.pyplot as plt
import numpy as np

# Fungsi keanggotaan fuzzy untuk Kecepatan Pelayanan
def pelayanan_lambat(x):
    return np.maximum(0, np.minimum(1, (7 - x) / 6))

def pelayanan_cepat(x):
    return np.maximum(0, np.minimum(1, (x - 1) / 6))

# Fungsi keanggotaan fuzzy untuk Kualitas Makanan
def makanan_buruk(x):
    return np.maximum(0, np.minimum(1, (7 - x) / 6))

def makanan_baik(x):
    return np.maximum(0, np.minimum(1, (x - 1) / 6))

# Fungsi keanggotaan fuzzy untuk Suasana Restoran
def suasana_buruk(x):
    return np.maximum(0, np.minimum(1, (7 - x) / 6))

def suasana_baik(x):
    return np.maximum(0, np.minimum(1, (x - 1) / 6))

# Variabel domain
x_pelayanan = np.linspace(1, 7, 100)
x_makanan = np.linspace(1, 7, 100)
x_suasana = np.linspace(1, 7, 100)

# Membuat plot
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Grafik Kecepatan Pelayanan
axs[0].plot(x_pelayanan, pelayanan_lambat(x_pelayanan), 'b-', label='lambat')
axs[0].plot(x_pelayanan, pelayanan_cepat(x_pelayanan), 'orange', label='cepat')
axs[0].legend()
axs[0].set_title("Kecepatan Pelayanan")
axs[0].set_xlabel("Kecepatan Pelayanan")
axs[0].set_ylabel("Derajat Keanggotaan")

# Grafik Kualitas Makanan
axs[1].plot(x_makanan, makanan_buruk(x_makanan), 'g-', label='buruk')
axs[1].plot(x_makanan, makanan_baik(x_makanan), 'r-', label='baik')
axs[1].legend()
axs[1].set_title("Kualitas Makanan")
axs[1].set_xlabel("Kualitas Makanan")
axs[1].set_ylabel("Derajat Keanggotaan")

# Grafik Suasana Restoran
axs[2].plot(x_suasana, suasana_buruk(x_suasana), 'purple', label='buruk')
axs[2].plot(x_suasana, suasana_baik(x_suasana), 'brown', label='baik')
axs[2].legend()
axs[2].set_title("Suasana Restoran")
axs[2].set_xlabel("Suasana Restoran")
axs[2].set_ylabel("Derajat Keanggotaan")

plt.tight_layout()
plt.show()
```
